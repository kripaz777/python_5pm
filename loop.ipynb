```python
# # WAP to create a billing application
# name of product
# price
# quantity
# total

# grand total
```

```python
bill = ""
grand_total = 0
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    price = int(input("Enter price = "))
    quantity = int(input("Enter quantity = "))
    total = price*quantity
    grand_total = grand_total + total
    data = name +" " + str(price) + " " +str(quantity) + " " +str(total) +"\n"
    bill = bill + data
    
print(bill)
print("All Total",grand_total)
```

```python
# prime - 2,3,5,7,11,13,17,19
# composite - 4,6,8,9,10,12,14

n = int(input("Enter n = "))
if n >2:
    for i in range(2,n):
        if n % i ==0:
            print("The no is composite")
            break
    else:
        print("This is prime")
        
else:
    print("The no should be >= 2")
```
